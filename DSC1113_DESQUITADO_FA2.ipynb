```python
import pandas as pd                   # For the dataframe structure
from functools import reduce          # For the reduce function

path = "/content/vaccinations.csv"    # The vaccination.csv dataset

df = pd.read_csv(path)                # The command used to read the csv file into the dataframe
df.head()                             # The command used to print the first 5 rows of the dataframe
```

```python
df['date'] = df['date'].map(pd.to_datetime)   # Command used to transform the 'date' to datetime format
df.head()
```

```python
rowsFiltered = list(filter(lambda row: row['people_fully_vaccinated_per_hundred'] > 70,     # Insert the filtered rows into a list
                           df.to_dict(orient = 'records')))                                 # The filtered rows are transformed first to dictionary so the insert values are in dictionary form
dfFiltered = pd.DataFrame(rowsFiltered)                                                     # This transforms the list into a dataframe
```

```python
def add(x, y):        # This is a function used to add two values
  return x + y

totalVaccinated = dfFiltered['total_vaccinations'].dropna().apply(lambda x: int(x))     # Removes the NaN values and formats the numerical values to integer (non-float)
totalVaccinatedCount = reduce(add, totalVaccinated)                                     # It adds the values together of sum and new value in complete iteration
print(totalVaccinatedCount)                                                             # Prints the result of the reduce
```

# The total number of vaccinated population who has more than 70 per hundred vaccinated is **$17,716,869,904,626$**.

Functional Programming mainly focuses on creating and formulating functions that would suffice in performing tasks, especially with large datasets that would take a great deal of time if not designed to be quick or instantaneous. This approach is solely concerned with the outcome or result, whether it meets the desired or needed.

Meanwhile, Imperative Programming gives great importance on procedure, making sure that the process is meticulously followed and clear. This could be beneficial for a programming compiled to be used and debugged by non-programming professional, such as administrative and managerial roles.

However, the length needed for outlining a clear course of action could affect the runtime of the project, thus slowing down the estimated runtime of the program as compared to the Functional program.

Functional Programming is beneficial for data processing and transformation as the size of the data itself can be a factor to consider when designing a system. If the data is considered big, the processing of the said dataset could consume a large amount of time, while also factoring in the hardware used in the process. On the other hamd, if we keep the functions simple and quick, the runtime could be slimmed down to an ideal duration.
